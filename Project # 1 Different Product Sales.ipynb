# Exploratory Data Analysis (EDA)

**Finding top selling product**

**Finding How many orders made by the coustmer by using group by**

```python
top_selling_product = df.groupby('PRODUCTCODE')['QUANTITYORDERED'].sum().sort_values(ascending=False).head(10)
top_selling_product
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(50, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=top_selling_product.index, y=top_selling_product.values, palette='viridis')
plt.title('Top Selling Products by Quantity',fontweight='bold')
plt.xlabel('Product Code' ,fontweight='bold')
plt.ylabel('Quantity Sold',fontweight='bold')
sns.despine()
plt.show()
```

**Finding top selling product**

```python
top_selling_product = df.groupby('PRODUCTCODE')['Profit'].sum().sort_values(ascending=False).head(10)
top_selling_product
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(50, 6))
plt.subplot(1, 2, 1)
sns.barplot(x=top_selling_product.index, y=top_selling_product.values, palette='cividis')
plt.title('Top Selling Products by Quantity',fontweight='bold')
plt.xlabel('Product Code' ,fontweight='bold')
plt.ylabel('Quantity Sold',fontweight='bold')
sns.despine()
plt.show()
```
