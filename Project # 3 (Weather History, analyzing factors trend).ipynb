**ther is no null value**

```python
df['Formatted Date'] = df['Formatted Date'].str.replace(' \+\d{4}', '', regex=True)

# Convert 'Formatted Date' to datetime format
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], format='%Y-%m-%d %H:%M:%S.%f')
df.head(4)
```

```python
df.info()
```

```python
# Create 'Date' and 'Time' columns
df['Date'] = df['Formatted Date'].dt.date
df['Time'] = df['Formatted Date'].dt.time
# Drop the original 'Formatted Date' column if needed
# df = df.drop(columns=['Formatted Date'])
df.head(2)
```

```python
df = df[['Date', 'Time'] + [col for col in df.columns if col not in ['Date', 'Time']]] #changing location of date and time columns
df.head(2)
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# Extract month from the 'Date' column and create a new column 'Month'
df['Month'] = df['Date'].dt.month
# Map month numbers to month names
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
               7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

df['Month'] = df['Month'].map(month_names) #adding month column
df['Year'] = df['Date'].dt.year  #adding year column

df = df[['Year'] + [col for col in df.columns if col not in ['Year']]] #changing the position of year at 1st column
df = df[['Month'] + [col for col in df.columns if col not in ['Month']]]


df.head(5)
```

# Exploratory Data Analysis (EDA)

**Temperature Trend** #making groups of temperature by groupby

```python
Year_Temperature_trend = df.groupby('Year')['Temperature (C)'].mean()
Year_Temperature_trend
```

```python
Month_Temperature_trend = df.groupby('Month')['Temperature (C)'].mean()
Month_Temperature_trend
```

**Humidity Trend**   #making groups of temperature by groupby

```python
Year_Humidity_trend = df.groupby('Year')['Humidity'].mean()
Year_Humidity_trend
```

```python
Month_Humidity_trend = df.groupby('Month')['Humidity'].mean()
Month_Humidity_trend
```

**Wind Speed**

```python
Year_WindSpeed_trend = df.groupby('Year')['Wind Speed (km/h)'].mean()
Year_WindSpeed_trend
```

```python
Month_WindSpeed_trend = df.groupby('Month')['Wind Speed (km/h)'].mean()
Month_WindSpeed_trend
```

**Pressure Trend**

```python
Year_Pressure_trend = df.groupby('Year')['Pressure (millibars)'].mean()
Year_Pressure_trend
```

```python
Month_Pressure_trend = df.groupby('Month')['Pressure (millibars)'].mean()
Month_Pressure_trend
```

**Visibility**

```python
Year_Visibility_trend = df.groupby('Year')['Visibility (km)'].mean()
Year_Visibility_trend
```

```python
Month_Visibility_trend = df.groupby('Month')['Visibility (km)'].mean()
Month_Visibility_trend
```
